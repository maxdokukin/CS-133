## **Question 4**: What is the difference between loc and iloc? When might we want to use one over the other?

The difference between iloc and loc is ...
* iloc is locating entry by the absolute index. for example, first entry will always have index 0
* loc will locate entry by its set index

iloc is more convinient when we want n_th item in the dataset. loc is more convinient if we want to get item with a specific to dataset index

 First make a dataframe of all the reasons people provided (their answers to this question). True's are considered a "provided reason."

 Then figure out the number of times each provided reason appears. (Series, DataFrame, whatever)

 Show me that (display it), then plot in part b.

````python
# Your answer here...

reasons = coffee[['Why do you drink coffee? (It tastes good)',
                  'Why do you drink coffee? (I need the caffeine)',
                  'Why do you drink coffee? (I need the ritual)',
                  'Why do you drink coffee? (It makes me go to the bathroom)']].astype('float').astype('Int64')

# reasons['Why do you drink coffee?'].sum(skipna=True)
# reasons['Why do you drink coffee? (It tastes good)'].astype('Int64')

reason_counts = {}
for column in reasons.columns:
  reason_counts[column.removeprefix('Why do you drink coffee? ')] = reasons[column].sum()

print(reason_counts)
````
